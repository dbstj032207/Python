## 2)지역별(구별)로 주유소 정보를 엑셀 파일로 저장하기

```python
from tqdm import notebook
import time
```

```python
for gu in notebook.tqdm(gu_names):
    tag = driver.find_element(By.ID, "SIGUNGU_NM0")
    tag.send_keys(gu)
    
    time.sleep(1)
    
    ##### 조회 버튼 누르기
    driver.find_element(By.XPATH, """//*[@id="searRgSelect"]/span""").click()
    
    time.sleep(1)
    
    ##### 엑셀 저장버튼 누르기
    driver.find_element(By.XPATH, """//*[@id="glopopd_excel"]/span""").click()
    
    time.sleep(1)
    
driver.close()
```

## (2) Data Definition

```python
from glob import glob
```

```python
station_files = glob("data/oil/지역*.xls")
```

```python
tmp_raw = []
for f in station_files:
    tmp = pd.read_excel(f, header=2)
    tmp_raw.append(tmp)
```

```python
len(tmp_raw)
```

```python
##### 데이터 프레임으로 합치기

station_raw = pd.concat(tmp_raw)
station_raw.head(50)
```

```python
station_raw.info()
```

## (3) Data Preprocessing

```python
##### 원하는 컬럼만 추출

stations = pd.DataFrame({"Oil_Store":station_raw["상호"], "주소": station_raw["주소"],
             "가격": station_raw["휘발유"], "셀프": station_raw["셀프여부"],
             "상표": station_raw["상표"]})
stations.head()
```

```python
##### 구 파생변수 추가
gu_list = [addr.split()[1] for addr in stations["주소"]]
stations["구"] = gu_list
```

```python
stations.head()
```

```python
print(stations["구"].unique())
print(len(stations["구"].unique()))
```

```python
##### 가격 결측치

stations["가격"].unique()
```

```python
stations[stations["가격"] == "-"]
```

```python
stations = stations[stations["가격"] != "-"]
```

```python
stations.info()
```

```python
##### 가격 정보를 수치로 변환
stations["가격"] = stations["가격"].astype(float)
```

```python
stations.info()
```

```python
##### 인덱스 재조정
stations.reset_index(inplace=True)  # drop=True
```

```python
stations.head(10)
```

```python
del stations["index"]
```

+ 주유 가격 상하위 10개씩을 지도에 표시

```python
oil_price_top10 = stations.sort_values(by="가격", ascending=False).head(10)
oil_price_bottom10 = stations.sort_values(by="가격", ascending=True).head(10)
```

```python
# 상위 10개 
# 구글 API 이용

lat_list = []    # 위도
lng_list = []    # 경도

for addr in notebook.tqdm(oil_price_top10["주소"]):
    
    temp = gmaps.geocode(addr, language="ko")

    temp_loc = temp[0].get("geometry")

    lat_list.append(temp_loc["location"]["lat"])
    lng_list.append(temp_loc["location"]["lng"])
    
oil_price_top10["lat"] = lat_list
oil_price_top10["lng"] = lng_list

map = folium.Map(location=[37.497955, 127.027619], zoom_start=11)


for n in oil_price_top10.index:
    folium.Marker(location=[oil_price_top10["lat"][n], 
                            oil_price_top10["lng"][n]],
                     icon=folium.Icon(color="red", icon="star")).add_to(map)

map
```

```python
# 하위 10개 
# kakao API 이용

import re

lat_list = []    # 위도
lng_list = []    # 경도

header = {"Authorization":"KakaoAK be9c427709d98b78ab9a851e8fb2ba4a"}
url = "https://dapi.kakao.com/v2/local/search/keyword.json?query='중부경찰서'"

for addr in notebook.tqdm(oil_price_bottom10["주소"]):

    addr = re.sub(pattern=r'\([^)]*\)', repl='', string=addr)
    url = "https://dapi.kakao.com/v2/local/search/keyword.json?query={}".format(addr)
    place = requests.get(url, headers=header).json()["documents"]
    
    lat_list.append(place[0].get("y"))
    lng_list.append(place[0].get("x"))
    

oil_price_bottom10["lat"] = lat_list
oil_price_bottom10["lng"] = lng_list

map = folium.Map(location=[37.497955, 127.027619], zoom_start=11)


for n in oil_price_bottom10.index:
    folium.Marker(location=[oil_price_bottom10["lat"][n], 
                            oil_price_bottom10["lng"][n]],
                     icon=folium.Icon(color="red", icon="star")).add_to(map)

map
```
